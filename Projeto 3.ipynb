```python
hourly_arrest_rate= policia[['stop_time', 'is_arrested']].groupby('stop_time').agg('mean', numeric_only=False)
hourly_arrest_rate.rename(columns={'is_arrested': 'hourly_arrest_rate'}, inplace=True)
hourly_arrest_rate.reset_index(inplace=True)
hourly_arrest_rate
```

<div class="alert alert-block alert-info">
3º QUESTÃO - C
</div>

```python
hourly_arrest_rate['stop_time'] = pd.to_datetime(hourly_arrest_rate['stop_time'])
g1 = hourly_arrest_rate.groupby(hourly_arrest_rate['stop_time'].dt.hour)['hourly_arrest_rate'].sum()
g1
```

```python
policia2 = policia.reset_index()
policia2 = pd.concat([policia2, g1], axis=1)
policia2.set_index('stop_datetime', inplace=True)
policia2
```

<div class="alert alert-block alert-info">
3º QUESTÃO - D
</div>

```python
g1.plot.bar(x='stop_time', y='hourly_arrest_rate', title='Taxa de Prisões por Hora do Dia')
```

```python
policia['stop_date'] = pd.to_datetime(policia['stop_date'])
drug_rate = policia.groupby(policia['stop_date'].dt.year)['drugs_related_stop'].sum()
drug_rate.plot()
```

```python
Annual_search_rate = policia.groupby(policia['stop_date'].dt.year)['search_conducted'].sum()
Annual_search_rate

Annual_drugsearch_rate = pd.concat([drug_rate, Annual_search_rate], axis=1)
Annual_drugsearch_rate.plot()
```
