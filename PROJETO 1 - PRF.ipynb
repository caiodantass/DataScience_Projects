```python
# Dados de 2018

dados_2018['mes'] = pd.to_datetime(dados_2018['data_inversa']).dt.month
```

```python
# Dados de 2018

mes_2018 = dados_2018.loc[:,['mes','pessoas','ilesos','feridos_graves']]
k = mes_2018.groupby(by='mes').sum()
prop_ilesos = (k['ilesos']/k['pessoas'])
prop_feridos = (k['feridos_graves']/k['pessoas'])
prop = pd.DataFrame(prop_ilesos).rename(columns = {0:'Proporção de ilesos'}).round(3)
prop['Proporção de feridos graves'] = pd.DataFrame(prop_feridos).round(3)
prop
```

```python
# Dados antes da pandemia
antes_da_pandemia = len(pd.concat([dados_2018,dados_2019]))

# Dados durante a pandemia
durante_a_pandemia = len(pd.concat([dados_2020,dados_2021]))

# Media dos anos
antes = antes_da_pandemia/2
durante = durante_a_pandemia/2

# Diferença das médias
diferença_pandemia = antes - durante
```

```python
df_pandemia = pd.DataFrame([antes,durante],index = ['Antes da Pandemia','Depois da Pandemia']).rename(columns = {0:'Media de incidência de acidentes '})
df_pandemia
```

```python
diferenca = pd.DataFrame([diferença_pandemia],index = ['Diferença entre os períodos']).rename(columns = {0:"Valor da diferença"})
diferenca
```

Com isso, é perceptível que, houve uma dimunuição da incidência de acidentes nas rodovias federais por causa da pandemia.

- Nessa situação, basicamente eu juntei os anos de 2018 e 2019 usando a função "concat", pois esses dois anos foram os que não tinham pandemia ainda, e depois juntei os anos de 2020 e 2021 também usando a função "concat", pois esses dois anos foram os que passaram pela pandemia, logo em seguida, como tinha valores de dois anos nas duas situações eu fiz a media de cada o dos dois, pois os diferenciei com as variaveis "antes_da_pandemia" e "durante_a_pandemia", e depois fiz o cálculo da diferença entre eles, e com a função "rename" mudei os nomes tanto das linhas como da coluna.
